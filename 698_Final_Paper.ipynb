#### **Abstract**

With the surging popularity of NLP models in the public domain, there will be greater expectations for applications that can seamlessly handle and respond to written user prompts. This research paper seeks to build and test various machine learning alternatives that can properly identify named entities for college basketball and classify those proper nouns to a specified group of labels (team, player, coach, and conference).
The data used to train the models was acquired from Reddit using a specific thread geared towards college basketball. The text prompts were then labeled and tokenized as each root word needs to be represented as a vector for further evaluation in a model. The two types of models evaluated in this research were larger neural networks trained on large corporas that would be fine tuned and a vectorized version of the labeled corpora that could be analyzed with an SVM, Multinomial Bayes, and Regression classifiers. Only a larger model was used to classify the named entities as there are complex patterns and grammar to be considered in order to properly identify these proper nouns. While each model was reasonably able to predict the classification labels, the Distilbert language model outperformed all other techniques at 77%; however, it was unable to reliably identify the tags for named entity recognition beyond 50%. Despite the slightly better performance, the more traditional machine learning techniques had reasonably close metrics (~76)  that might justify using a simpler approach to run classification tasks like this one.

### **Literature Review**

### Named Entity Recognition

Models that can properly recognize relevant entities have always been sought after as they are the foundation for breaking text down and beginning to convert this data into a more structured understandable format. It can substantially improve the ability of a model to hone in on the key subjects of a sentence or document, which can further be used in alternative contexts such as document summarization and text similarity comparisons. Pascale (2023) reviewed the wide variety of literature that existed for NER which provided helpful historical context of the variety of methods available to complete this task. The earliest attempts to identify named entities within text involved building rules (typically using regular expression) based on available corpora of text data. This method of identification of named entities can become quite cumbersome as new rules have to be built to correctly account for varieties in human based search. Eftimov T, Koroušić Seljak B, Korošec P (2017) indicates the rules approach can be highly domain specific, which means that this process would need to be repeated for each new context or topic area. In standardized forms of literature it may be good enough to apply these rules, but the modern digital world has considerable variety in terms of how humans write and communicate online that is far different from historical written works. Groups of words can often help to identify named entities and it can be very difficult to build rules from scratch that account for the considerable variety that is common across different authors.  Eftimov T, Koroušić Seljak B, Korošec P (2017) also discuss that more modern machine learning methods can be very time consuming to develop as they require labeled annotated corpora to evaluate the accuracy of these models. In some contexts, they also require human expertise in a subject matter to properly identify the named entities in a phrase or document. These models would subsequently build generic rules based on numeric representations of the text that could be used rather than manually pulling together the different patterns that exist in text. Lastly, there are unsupervised methods that would be prompted with relevant entities that can learn the rules on a training set which is subsequently applied to new unseen data. The downside with this method is that there could be new means of online communications used that a model might need to be retrained on to accurately identify named entities; however, the systems being used in this context have been trained on very large corpora which allow them to recognize many patterns of human text beyond specific words. The approach taken with this research project was to provide a corpora based on human labeled Reddit comments on college basketball which will then be used to evaluate modeling techniques.

#### Modern NER Techniques

Pascale (2023) listed some of the more modern techniques for NER, which were primarily variants of different neural networks (RNN, CNN,LSTM).  Each neural network variant was initially constructed to try to solve a slightly different aspect of written texts. The primary disadvantage of these methods is that they are designed to evaluate the relationships between subsets of words. One somewhat larger limitation related to these modeling techniques is how they track and evaluate the numeric representations of words and series of words. A vector is created as an embedding of a word and is evaluated in a sequential order. This is typically in a forward direction given the structure of sentences although backward evaluations can also be applied. One method will be tested that relies on such a highly trained long short term memory (LSTM) with some of the state of the art training techniques that seek to maximize model performance. Vaswani et al (2017) identified a new mechanism, attention, that allowed for training evaluation to consider the relationships of words in a text and the interplay between them rather than a more linear or sequential pattern as what typically was done in other state of the art NLP models. This numeric weight helped to minimize some of the terms that were forgotten by these models as they progressed through many hidden layers. Another large disadvantage prevalent to all neural network NLP models is that given the variety of ways to describe the same object in written text the models store many parameters that must be repeatedly trained. This typically requires more powerful processing RAM in order to even fine tune a pre-trained model.
Tunstall (2022) indicates the largest difference that exists in the initial transformer models is the concept of self-attention. Weights for hidden states are all computed together rather than sequentially or in a specific order. The embedding that is a representation of the text data in a vector uses a weighted average of each embedding rather than one fixed vector that is identified for each and every tokenized word. The embeddings/attention scores are calculated using a similarity function which is a dot product of the embeddings found in the hidden state. The more common the overlap between words in specific statements the more likely the model will expect a specific word or phrase in relation to that current word. This modification led to the creation of BERT and GPT variants that were earlier siblings with much smaller training sets of the much hyped large language models that have grabbed the attention of many individuals and corporations.

#### Transfer Learning

The concept of transfer learning has long been common practice within the machine learning community for image analysis and is becoming a far more popular strategy when running NLP modeling as well. The massive size of NLP datasets and trained models allow others access to state of the art models without having to run as much computationally expensive training to build successful predictive models. Despite being able to utilize these previously trained models, their size remains large and complex which are often trained for general contexts. Pascale (2023) reviews some of the more fine-tuned models that exist in finance, biology, and legal contexts which makes sense given specific terminology that exist in those domains. For others that don’t want to spend time building a larger model like those referenced in the paper, it is possible to fine tune more standardized models with two different methods. Tunstall (2022) provides an overview for transformer models that include fine tuning the classification head that relies on the pre-trained weights that exist in publicly available models. The performance of said models can vary depending on each use case, but it is an easier way to analyze and predict data particularly without more advanced chip resources. The alternative case covered by Tunstall is a more thorough refactoring of the underlying neural network that allows the weights and attention methods to be fully adjusted to a specific corpora. This latter method is a more comprehensive approach to leverage these models, but requires more resources to properly compute given the size of even earlier generation models.
Other researchers have also experimented with techniques to maximize the utility of transfer learning. One example that was utilized in this project based on the research from Howard and Ruder (2018) indicates that some of the methods to maximize the benefits of fine tuning can substantially improve performance. The primary ways that are discussed in their paper are based on modifying learning rates by initially ramping up and then subsequently decreasing the rate.. These gradient differences also imply that there are varied levels of training for each layer within a neural network which is effective given that different underlying patterns are calculated by the weights of a given layer. The other method discovered to be very effective particularly for layered models was to gradually make different layers available for updates. The authors call this unfreezing and eventually make all of the layers available for retraining. All of these ways in combination were found in their research to fine tune a large model that is performant for a wide variety of NLP tasks.

#### TF-IDF

One modeling technique that preceded embeddings that existed in neural networks as a technique to identify text patterns is TF-IDF. The premise for the model takes the inverse of each term’s frequency to find the ratio between how many documents a term appears compared to the total documents in the sample. Sarkar (2016) indicates it is designed to more prominently weight rarer terms that have more significance compared to words that appear in many documents. This creates many numeric features that represent the patterns of words that exist in a corpora. This technique allows for a wide variety of supervised classification models to be applied as they can learn the mixture of words and their frequency that most closely align with different labels. One disadvantage is that its corpora is more limited to the training set and may not generalize well to very different data prompts.

#### Sparse Matrix Classification

There are a number of well known classification techniques that are well suited to handle the sparse matrix that is created from the TF-IDF encoding process. Sarkar (2016) mentions SVM and Multinomial Bayes (MNB) as two potential preferential available models that can handle the large dimensionality that exists. The author also indicates that MNB is more of a benchmark approach to identify if the SVM model is statistically more effective. Kuhn & Johnson (2018)  discuss the traditional usage for SVM models in a linear context particularly for linear datasets and relationships; however, there are a range of functions that can be applied to properly account for nonlinear relationships that exist in datasets as would typically be expected from TF-IDF matrices. Sarkar (2016) discusses how it is an effective classifier as it can identify the hypothetical hyperplane that represents the space in multidimensional vectors between each of the different labeled classifications. Zheng (2015) indicates that logistic regression is a simple yet effective model to evaluate against TF-IDF as well given its ability to use the sigmoid function as a means of classification. The author goes on to discuss regularization options as a means of improving performance given that a sparse matrix has a larger number of potential statistically insignificant predictors that can be excluded.

### **Methodology**

#### Data Collection

Given the substantial nuance within written text it did not seem plausible to train a model from scratch. As indicated in the prior literature review, a model would need to be heavily trained on a massive corpus of text data to account for the substantial variety of written words that exists in text data. Therefore, the approach of this research was to evaluate several  methods using transfer learning and fine tuning that could be applied to the specific focus area on college basketball data.


The data itself was not so difficult to acquire as there are digital forums that allow for API consumption of user commentary. The website used has subthreads that are meant to focus on specific topic areas that help to organize users based on their interest in specific subjects. This structure makes it easier to collect similarly themed text corpora that could be used within this research project to train models on very narrow topic areas. The records were human labeled to attempt to maximize the accuracy of the potential classification that could be ingested in the model although there were some errors that could still be made given the sheer volume of labels that were needed. A relative assisted with some of the labeling and needed to search online or ask about the named entities in specific posts. The commentary itself could have a wide variety of topics within college basketball, but also had many other references to pop culture. Users could have typos and other grammatical errors that might exist that wouldn’t follow expected conventions or proper grammar to discuss the college basketball affiliated entities. The posts were left as originally written on the online forum and users could choose to include additional content that did not directly relate to the subject of the post including images, gifs, and other content that was out of scope from this analysis; however, in general it was more homogenous content that could be tested against different model types.

#### Data Preparation

Although the API provided an easy means for acquiring the data needed for this research project, there was still preparation and clean up work needed to get the corpora in usable formats for subsequent modeling. The approach recommended and followed by the state of the art models is to tokenize text data, which can involve several different steps. The models utilizing TF-IDF vectors develop features from individual words or phrases that have already been cleaned of punctuation and other extraneous information in the posts. Many reddit posts included emojis and links that would be excluded as inputs as there were limited insights that can be extracted from them for the scope of this project. After evaluating the number of terms that would be created with bigrams, it seemed as though there were far too many combinations of two plus words in the entire corpora.
The following steps were taken to prepare the data:


##### Tokenization

The first step to prepare the data involved looping through each post and separating every word into a list. There are a number of ways to break apart the words from one another although there are some common Regex patterns that have been adopted by natural language processing packages in Python that are fairly effective at completing this task. Larger neural network models, such as ULMFIT  and DISTILBERT require using the exact tokenizer that trained the initial model as it ensures the inputs are most closely aligned with the prior corpora to maximize its performance. There are slightly different formats for each given model, but the neural networks primarily include certain prefixes to denote the beginning of a sentence and sometimes split certain rare words apart.

The remaining preprocessing is only required for models that utilize TF-IDF encoding.


#### Removing Special Characters and Stopwords

It is beneficial to remove any punctuation and unusual characters that may exist in the Reddit posts as the TF-IDF approach is based on the frequencies of words across all of the different written prompts. Although these types of inputs may be helpful for sentiment analysis models given their impact on tone and intent for the purposes of this classification they do not add much value and are excluded from the corpora. Linguists have also put together a list of words that do not have much meaning at all such as “the”, “a”,”that”, “if”. The word “the” is the most frequent word that exists in English texts, but does not hold much significance.

##### Lemmatization

This is an optional step that was chosen to reduce the corpora size and involves breaking words down into their root form. There are many different variations of the same word due to conjugation, tense, and sentence construction that ultimately have the same meaning. An example of a group of words that would be modified into the same form after lemmatization would be “change”, “changed”,”changes”, and “changing”. The lemmatization process attempts to further minimize the corpora size by eliminating the variation that exists of the same word and identify the underlying relationships of these words to the classification labels. There are potential risks with this approach of losing potential meaning, but ultimately the underlying root word is more valuable to improve model performance than slightly different variation of the same word.


##### TD-IDF Encoding

Once the text strings are broken up and lemmatized into a more uniform corpus, the vectorizer is applied to the entire corpus from the training set. This function first uses a bag of words model and obtains a count of each unigram that exists in the corpora across the set and represents that information in a sparse matrix for each identified token. Next, a statistic is calculated that is more beneficial for inference by determining how prevalent the term is within a specific text prompt versus the number of times the term appears in all of the prompts.

### **Conclusion & Future Areas for Exploration

The primary aim of this research was to evaluate the different available NLP techniques that could classify text prompts and identify the named entities and their respective labels. From looking through other literature and discussion around state of the art models it was clear that there were two main approaches that were recommended by others in the field. The first was to try to fine tune models that have been trained on large corpora and apply transfer learning on the narrowed scope of a new training set. This technique could be used on both named entity recognition and labeled classification as the immense training sets allows these architectures to treat both tasks as a classification exercise. The second was to use more traditional machine learning classifiers based on techniques to vectorize the training corpora and identify patterns between the tokenized words.


The number of Reddit posts were queried and labeled in order to be used for training and test sets for this analysis. The prompts needed slightly different preparation techniques depending on the model used for classification, but all of the prompts need to be tokenized into separate words to be properly ingested for further processing. The neural networks took care of the embedded tokens and expected weights to be used within their models; while other machine learning techniques needed lemmatization and cleaning of non-words and stopwords before creating numeric representations of the text.
The first task of trying to identify named entities proved to be rather challenging due to the fact that the Reddit posts used many different acronyms as shorthand for each entity. The scope of this analysis was also very focused on a subset of potential labels and there were often references to other proper nouns that a typical named entity recognition model would expect to identify from a prompt. Many of the names of teams correspond with US states and the prompts would interchange the references between the location itself and the organization that indirectly represents that location. Given that nuance in the data the model was able to classify around 50% of the tags for named entities, which is far below an acceptable level of performance to scale it to any future data.


There were a number of models evaluated to see how well each one could classify that label associated with the named entities. The data was stratified given the imbalance of the labels in the training and test set; however, it appeared that there was an impact on the classification given that the highest frequency label had a much higher false positive rate than any of the other labels. One of the highly trained models performed rather poorly overall without classifying two of the labels correctly, while the other larger neural network had the best performance of any model reviewed in this research. The classical machine learning classifiers (SVM, Logistic Regression, and Ridge Regression) performed on par with the bigger language models while only slightly compromising on accuracy and F1 metrics.
Ultimately, given the fact that many of the models returned fairly similar results it begs the question if it is necessary to use more powerful resource intensive models without a requisite bump in performance. Given the added cost and computation time of the larger language models even with far less parameters than the most popular public versions today it does not seem as though they provided great return on investment unless the tasks involve more complex semantic processing. The deep understanding that is developed on written text is absolutely valuable, but should be applied on a case by case basis that leverages the power of these models to accomplish tasks that other machine learning techniques have limited ability to accomplish.

#### Future Areas of Research
Given the previously discussed limitation and the specificity that exists within this subject area a future research area would involve evaluating vector similarities of different posts in a vector database. This newer database architecture may allow a system or model to more seamlessly evaluate prompts or queries and return similar results to meet user needs; however, it begs the question of what technique is best to properly represent the vectors so that a database could properly calculate similarity statistics. It appears that there will still be some reliance on creating high quality vectors or embeddings that accurately represent the text that likely would have to utilize the models discussed in the research.


Another area for future review would be to consider unsupervised techniques that would feed lists of named entities that could try to match those examples in unlabeled text. The time spent labeling was very intensive and it would be interesting to see if an unsupervised technique could be applied that would save the requisite time and perform at comparable levels of accuracy.
